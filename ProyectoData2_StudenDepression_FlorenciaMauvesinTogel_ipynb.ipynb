Descripciones de los Campos

* id:
Un identificador único asignado a cada registro de estudiante en el conjunto de datos.

* Género (Gender):
El género del estudiante (p. ej., Masculino, Femenino, Otro). Esto ayuda a analizar las tendencias de salud mental específicas por género.

* Edad (Age):
La edad del estudiante en años.

* Ciudad (City):
La ciudad o región donde reside el estudiante, proporcionando contexto geográfico para el análisis.

* Profesión (Profession):
El campo de trabajo o estudio del estudiante, que puede ofrecer información sobre los factores de estrés ocupacional o académico.

* Presión Académica (Academic Pressure):
Una medida que indica el nivel de presión que enfrenta el estudiante en entornos académicos. Esto podría incluir el estrés de los exámenes, las tareas y las expectativas académicas generales.

* Presión Laboral (Work Pressure):
Una medida de la presión relacionada con las responsabilidades laborales, relevante para los estudiantes que están empleados junto con sus estudios.

* CGPA (cumulative grade point average of the student):
El promedio general acumulado de calificaciones del estudiante, que refleja el rendimiento académico general.

* Satisfacción con los Estudios (Study Satisfaction):
Un indicador de cuán satisfecho está el estudiante con sus estudios, lo que puede correlacionarse con el bienestar mental.

* Satisfacción Laboral (Job Satisfaction):
Una medida de la satisfacción del estudiante con su trabajo o entorno laboral, si corresponde.

* Duración del Sueño (Sleep Duration):
El número promedio de horas que el estudiante duerme por día, lo cual es un factor importante en la salud mental.

* Hábitos Alimenticios (Dietary Habits):
Una evaluación de los patrones de alimentación y los hábitos nutricionales del estudiante, que potencialmente impactan la salud general y el estado de ánimo.

* Título / Nivel de Estudios (Degree):
El título académico o programa que está cursando el estudiante.

* ¿Alguna vez has tenido pensamientos suicidas? (Have you ever had suicidal thoughts ?):
Un indicador binario (Sí/No) que refleja si el estudiante alguna vez ha experimentado ideación suicida.

* Horas de Trabajo/Estudio (Work/Study Hours):
El número promedio de horas por día que el estudiante dedica al trabajo o al estudio, lo que puede influir en los niveles de estrés.

* Estrés Financiero (Financial Stress):
Una medida del estrés experimentado debido a preocupaciones financieras, que pueden afectar la salud mental.

* Antecedentes Familiares de Enfermedad Mental (Family History of Mental Illness):
Indica si existen antecedentes familiares de enfermedad mental (Sí/No), lo que puede ser un factor significativo en las predisposiciones de salud mental.

* Depresión (Depression):
La variable objetivo que indica si el estudiante está experimentando depresión (Sí/No). Este es el enfoque principal del análisis.

### Tratamiento de Outliers para la variable Age

```python
# Calcular los cuartiles para la columna 'Age'
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

# Definir el limite superior para los outliers (usando la regla 1.5 * IQR)
# Los outliers superiores son los valores que están por encima de Q3 + 1.5*IQR
upper_bound_age_outlier = Q3_age + 1.5 * IQR_age

# Identificar los valores atipicos superiores en 'Age'
outliers_age_data = df[df['Age'] > upper_bound_age_outlier]

print(f"--- Análisis de Outliers en 'Age' ---")
print(f"Cuartil 1 (Q1) de Age: {Q1_age}")
print(f"Cuartil 3 (Q3) de Age: {Q3_age}")
print(f"Rango Intercuartílico (IQR) de Age: {IQR_age}")
print(f"Límite superior para outliers en Age (Q3 + 1.5*IQR): {upper_bound_age_outlier:.2f}\n")

print(f"Cantidad total de outliers en 'Age': {len(outliers_age_data)} registros")
print(f"Porcentaje de outliers en 'Age' respecto al total del dataset: {(len(outliers_age_data) / len(df)) * 100:.2f}%")
```

```python
#Eliminacion de Outliers en 'Age' ya que representa solo un 0.04% del total.
# Guardamos la cantidad original de filas antes de la eliminación
initial_rows = len(df)

# Filtramos el DataFrame para conservar solo las filas donde 'Age' no es un outlier. Es decir, donde 'Age' es menor o igual al limite superior para outliers
df = df[df['Age'] <= upper_bound_age_outlier].copy()


# Volvemos a verificar la distribución de 'Age' después de la eliminación
plt.figure(figsize=(6, 1))
sns.boxplot(data=df, x='Age')
plt.title("Boxplot de 'Age' después de la eliminación de outliers")
plt.xlabel("Age")
plt.show()
```

## Análisis de Correlación


Después de explorar las relaciones entre variables específicas y la depresión a través de gráficos de conteo para nuestras hipótesis, es importante examinar las correlaciones numéricas generales en nuestro dataset. Este análisis nos permitirá identificar si las variables numéricas tienden a moverse juntas de forma lineal o monotónica, y cómo se relacionan con la variable objetivo, la depresión.

Comenzamos evaluando las correlaciones de Pearson, que miden la fuerza y dirección de las relaciones lineales. Esto nos daría una primera aproximación a la interdependencia entre nuestras variables.

```python
# Busco la matriz de correlacion de TODAS las variables numericas, incluida 'Depression'.
current_num_vars = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Calcular la matriz de correlacion
correlation_matrix = df[current_num_vars].corr()

# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap de Correlación de Pearson de Variables Numéricas', fontsize=11)
plt.show()
```

**Insigth**: Hay bajas Correlaciones Lineales. Al analizar el heatmap de Correlación de Pearson, notamos que la mayoría de los coeficientes de correlación entre nuestras variables numéricas y con la variable objetivo "Depression" son relativamente bajos, cercanos a cero. Pocos valores se acercan a 0.5 o -0.5, lo que sugiere que no existen relaciones lineales fuertes en nuestros datos.

Este hallazgo inicial me resultó un tanto sorprendente, dado que en nuestros barplots de hipótesis previas habíamos observado claras tendencias y proporciones que indicaban una relación entre variables como Sleep Duration, Financial Stress, Academic Pressure, Study Satisfaction y la Depression.



Como proximo paso explorare correlaciones monotónicas: La Matriz de Spearman

Ante las bajas correlaciones lineales mostradas por Pearson, y considerando la naturaleza de algunas de las variables (como Financial Stress, Academic Pressure, Study Satisfaction y Sleep Duration), que son ordinales aunque tratadas numéricamente (escalas de 1 a 5, o rangos de horas), voy a calcular como proximo paso, la correlación de Spearman.

Esta más adecuada para este tipo de variables (donde una variable siempre aumenta o disminuye con la otra, pero no necesariamente de forma lineal). Esto permitiría ver si, a pesar de la falta de linealidad estricta, existían tendencias consistentes en los rangos de los datos.

```python
# Probamos con Spearman
# Identificamos las variables numericas en el DataFrame 'df' actual.
current_num_vars = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Calcular la matriz de correlacion de Spearman
correlation_matrix_spearman = df[current_num_vars].corr(method='spearman')

# Generar el heatmap de Spearman
plt.figure(figsize=(6, 4)) # Ajusta el tamaño para una mejor visualizacion
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap de Correlación de Spearman de Variables Numéricas', fontsize=11)
plt.show()
```

**Insights:** El heatmap de Correlación de Spearman también nos confirmó que las correlaciones entre las variables predictoras numéricas y la Depression son, en su mayoría, bajas.

**Mapeo de Variables Nominales** One-Hot-Encoding

Evaluacion de RELEVANCIA de las Variables

```python
from scipy.stats import chi2_contingency

print("Evaluación de Relevancia de Variables Categóricas Nominales (Chi-Cuadrado)")

nominal_columns = ["City", "Profession", "Dietary Habits", "Degree"]
target_column = "Depression"

for col in nominal_columns:
    print(f"Analizando la asociación entre '{col}' y '{target_column}':")
    # Crear una tabla de contingencia
    contingency_table = pd.crosstab(df[col], df[target_column])

    # Realizar la prueba de Chi-cuadrado
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    print(f"  Valor Chi-cuadrado: {chi2:.2f}")
    print(f"  Valor p: {p_value:.4f}") # Formatear a 4 decimales para mayor claridad

    if p_value < 0.05:
        print(f"  --> Existe una asociación estadísticamente significativa entre '{col}' y 'Depression'.")
        print(f"  (Se recomienda incluir '{col}' en el modelo después del One-Hot Encoding).")
    else:
        print(f"  --> NO existe una asociación estadísticamente significativa entre '{col}' y 'Depression'.")
        print(f"  (Considerar excluir '{col}' del modelo para simplificación, o investigarla más a fondo).")
```

## Modelo de Clasificación: Random Forest

## Optimización de los hiperparámetros del modelo (Random Forest Classifier)

Nos enfocaremos en los parámetros que más impacto tienen en el sobreajuste (overfiting):

* n_estimators: El número de árboles en el bosque. Más árboles suelen mejorar el rendimiento, pero con rendimientos decrecientes y mayor tiempo de cómputo.
* max_depth: La profundidad máxima de cada árbol. Limitarla ayuda a reducir el sobreajuste.
* min_samples_leaf: El número mínimo de muestras que debe tener una hoja del árbol. Aumentarlo también reduce el sobreajuste.
* min_samples_split: El número mínimo de muestras requeridas para dividir un nodo interno.

```python
from sklearn.model_selection import GridSearchCV

# Definimos el rango de hiperparametros a probar
param_grid = {
    'n_estimators': [100, 200, 300], # Prueba 100, 200, 300 arboles
    'max_depth': [None, 10, 20],     # Prueba sin límite, 10 o 20 de profundidad
    'min_samples_leaf': [1, 5, 10],  # Prueba 1, 5 o 10 muestras minimas por hoja
    'min_samples_split': [2, 5, 10]  # Prueba 2, 5 o 10 muestras minimas para dividir
}

# Instanciamos el modelo Random Forest (con un random_state para reproducibilidad)
rf_base = RandomForestClassifier(random_state=42, class_weight='balanced')

# Configuramos GridSearchCV
# cv=5 significa 5-fold cross-validation
# scoring='roc_auc' elige la mejor combinacion de parametros
# basándose en el ROC AUC (por laclasificacion binaria)
# n_jobs=-1 para acelerar el proceso
grid_search = GridSearchCV(estimator=rf_base, param_grid=param_grid,
                           cv=5, scoring='roc_auc', n_jobs=-1, verbose=1)

print("Ejecutando la búsqueda en cuadrícula")
# Ejecuta la búsqueda en los datos de entrenamiento
grid_search.fit(X_train, y_train)

print("Optimización Completada")
print(f"Mejores hiperparámetros encontrados: {grid_search.best_params_}")
print(f"Mejor puntuación ROC AUC (validación cruzada): {grid_search.best_score_:.4f}")

# Obtenemos el mejor modelo
best_rf_model = grid_search.best_estimator_
print("El mejor modelo optimizado ha sido seleccionado.")
```

**Análisis de los Resultados de la Optimización:**
Fitting 5 folds for each of 81 candidates, totalling 405 fits: Se entrenaron y evaluaron 405 modelos.

**Mejores hiperparámetros encontrados:**

* max_depth: None: Es interesante que el modelo prefiera la profundidad máxima sin límite. Esto significa que los árboles individuales pueden crecer lo suficiente como para capturar patrones muy complejos. Sin embargo, la combinación con min_samples_leaf es clave.
* min_samples_leaf: 5: Este es el parámetro que realmente está ayudando a combatir el sobreajuste. Al requerir que cada hoja final tenga al menos 5 muestras, los árboles no se ajustan demasiado a puntos de datos individuales o ruido.
* min_samples_split: 2: Este es el valor por defecto y sugiere que la división de un nodo puede ocurrir incluso si solo hay 2 muestras en él, pero min_samples_leaf=5 actuará como un límite más estricto.
* n_estimators: 300: El modelo encontró que usar 300 árboles ofrece el mejor equilibrio entre rendimiento y tiempo de cómputo para este rango de parámetros. Más árboles generalmente resultan en un modelo más robusto.

**Mejor puntuación ROC AUC (validación cruzada)**: 0.9191
Esta es la puntuación promedio de ROC AUC obtenida por el mejor modelo en las 5 iteraciones de validación cruzada. Es un valor muy fuerte y ligeramente superior al 0.9111 que obtuve con el modelo no optimizado en el conjunto de prueba. Esto indica que la optimización fue buena en encontrar una mejor combinación de parámetros.
